### IsActiveMember

```python
base[(base["IsActiveMember"] == 0) & (base["Exited"] == 1 )]
```

1.302 membros (63.91%) que encerram a conta, já não eram membros ativos no banco. 

```python
def categoria_quali_quanti(coluna):
    colunabase = base.groupby([coluna, "Exited"]).size()
    coluna = base.groupby(coluna).size()
    return (colunabase / coluna*100).reset_index()
```
