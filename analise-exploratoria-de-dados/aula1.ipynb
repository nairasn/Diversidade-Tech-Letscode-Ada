```python
file = open('atividade_Arquivos.txt', 'r')
lista = file.readlines()
dict_nomes_comp = {}
ids = []
nomes = []
qtds = []
for line in lista[1:]:
    ids.append(line.strip('\n').split(';')[0])
    nomes.append(line.strip('\n').split(';')[1])
    qtds.append(line.strip('\n').split(';')[2])
    
dict_nomes_comp['id_municipio'] = ids
dict_nomes_comp['nomes'] = nomes
dict_nomes_comp['quantidade'] = qtds
```
